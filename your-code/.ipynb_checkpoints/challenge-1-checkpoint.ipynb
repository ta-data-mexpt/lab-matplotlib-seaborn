#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way
plt.subplot(1,2,1)
plt.plot(x,y,color="r")
plt.subplot(1,2,2)
plt.plot(x,z, lw = 4)
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)

fig,axes=plt.subplots(1,2)
axes[0].plot(x,y,color="r")
axes[1].plot(x,z,lw=4)
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(10,10))

axes[0].plot(x,y,color="red")
axes[1].plot(x,z,color="blue")

#dado que los ejes x tienen los mismos números decidí compartirlos entre las dos gráficas
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(10,10))

axes[0].plot(x,y,color="red")
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="blue")
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')

#sólo la gráfica de abajo necesita tener el nombre ya que es el mismo para ambas gráficas
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
#x = exp(x)
#y = x**2

fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```
