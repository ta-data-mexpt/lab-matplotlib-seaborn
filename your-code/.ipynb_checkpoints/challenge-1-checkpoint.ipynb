#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
plt.figure(figsize = (15, 5))
a = plt.subplot(1, 2 ,1)
b = plt.subplot(1, 2, 2)
a.plot(x, y, color = 'grey', linestyle = '-.')
a.set_xlabel('Eje X')
a.set_ylabel('Eje Y')

b.plot(x, z, color = 'pink', linestyle = '--')
b.set_xlabel('Eje X')
b.set_ylabel('Eje Y');
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.figure(figsize = (15, 5))

a = plt.subplot(1, 2 ,1)
b = plt.subplot(1, 2, 2)

a.plot(x, x**2, x, np.exp(x), color = 'grey', linestyle = '-.')
a.set_title('Normal Scale')

b.plot(x, x**2, x, np.exp(x), color = 'pink', linestyle = '--')
b.set_yscale('log')
b.set_title('Logarithmic Scale');
```

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

```python
vehicles=pd.read_csv('vehicles.csv')
vehicles.head()
```

## Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
sns.scatterplot( x = vehicles['Combined MPG'], y = vehicles['Highway MPG'], data = vehicles, alpha = 0.5)
plt.title('Combined MPG & Highway MPG', fontsize = 15);
```

## Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
sns.boxplot(y = vehicles['CO2 Emission Grams/Mile'])
plt.title('CO2 Emission Grams/Mile', fontsize = 15);
```

## Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
plt.hist(vehicles['Fuel Barrels/Year'], bins = 50)
plt.title('Fuel Barrels/Year', fontsize = 15)
plt.xlabel('Years')
plt.ylabel('Fuel Barrels');
```

## Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
plt.bar(vehicles['Fuel Type'], vehicles['City MPG'])
plt.xticks(rotation = 90)
plt.title('Fuel Type vs City MPG', fontsize = 15)
plt.xlabel('Fuel Type')
plt.ylabel('City MPG');
```
